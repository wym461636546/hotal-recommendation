```python
#define the display function
def get_hotels(df):
    df = df.rename(columns={'name': 'Hotel_Name'})
    print("Please tell us the following information:")
    destination = input("Which city will you travel to: ").lower()
    review_sentiment_input  = input("Would you like Positive or Negative Reviews: ").lower()
    category = input("Are you looking for reviews about certain topic? \nPlease select from this list {price, service，safety,cleanliness, atmosphere, food, location, room, facility, staff}: ").lower()
    category = [category.strip() for category in category.split(',')]
    review_sentiment = 1 if review_sentiment_input  == 'positive' else 0

    filtered_df = df[(df['city.name'].str.lower() == destination ) &
                     (df['category_label'].str.lower().isin(category)) &
                     (df['adjust_sentiment'] == review_sentiment)]

    grouped_hotel =  filtered_df.groupby(['Hotel_Name'])['review_rating_overall'].mean().reset_index()

    print('')
    print(f'We have found {len(filtered_df.Hotel_Name.value_counts())} different hotels that meet your expectations.\n')
    print('Please find 3 of them below:')
    if review_sentiment == 1:
        print(grouped_hotel['Hotel_Name'].tail(3))
    else:
       print(grouped_hotel['Hotel_Name'].head(3))

    print('')
    print(f'Please find the Hotels in {destination} and have {review_sentiment_input} reviews about {category}.\n')


    if review_sentiment == 1:
        top_hotels = filtered_df.sort_values(by='review_rating_overall', ascending= False)
        top_hotels = top_hotels[['Hotel_Name', 'review_comment', 'address','review_rating_overall','category_label']]
        display(top_hotels)
    elif review_sentiment == 0:
        bottom_hotels = filtered_df.sort_values(by='review_rating_overall', ascending= True)
        bottom_hotels = bottom_hotels[['Hotel_Name', 'review_comment', 'address','review_rating_overall','category_label']]
        display(bottom_hotels)
    else:
        print("Invalid review sentiment input. Please enter 'positive' or 'negative'.")
        return
```

```python
get_hotels(df = my_df)
```
